```python
# Open In Colab
# - Phase 01 -
# Installing Necessary Dependencies

# Step 0: Imports
from google.colab import files
import os, shutil, zipfile, random
from PIL import Image
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import cv2

# Step 1: Create folder structure for this dataset
!mkdir -p /content/data/panda /content/data/bear /content/data/logs

# Step 2: Upload the ZIP that contains Panda/Bear images (any reasonable folder layout is okay)
print("📤 Upload the ZIP containing Panda & Bear images (e.g., 'panda & bear.zip')...")
uploaded = files.upload()
zip_paths = [fn for fn in uploaded if fn.lower().endswith(".zip")]
if not zip_paths:
    raise ValueError("No ZIP uploaded. Please upload a .zip with panda and bear images.")
zip_path = zip_paths[0]
print(f"✅ Found ZIP: {zip_path}")

# Step 3: Unzip into a temporary directory
unzip_dir = "/content/dataset_raw"
os.makedirs(unzip_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zf:
    zf.extractall(unzip_dir)
print("✅ Unzipped.")

# Helper: basic image file check
def is_img(fn):
    fnl = fn.lower()
    return fnl.endswith((".jpg",".jpeg",".png",".bmp",".tif",".tiff",".webp"))

# Step 4: Find class folders that actually contain images
def find_class_dirs(root):
    pandas, bears = [], []
    for r, d, f in os.walk(root):
        img_count = sum(is_img(x) for x in f)
        if img_count > 0:
            dn = os.path.basename(r).lower().replace(" ", "_")
            if "panda" in dn:
                pandas.append(r)
            if "bear" in dn:
                bears.append(r)
    return sorted(pandas), sorted(bears)

panda_dirs, bear_dirs = find_class_dirs(unzip_dir)
if not panda_dirs or not bear_dirs:
    raise FileNotFoundError(
        "Couldn't detect both panda and bear image folders in the ZIP. "
        "Make sure folders are named with 'panda' and 'bear' in their names."
    )

print("🐼 Panda dirs (sample):", panda_dirs[:3], ("(+ more)" if len(panda_dirs)>3 else ""))
print("🐻 Bear  dirs (sample):",  bear_dirs[:3],  ("(+ more)" if len(bear_dirs)>3 else ""))

# Step 5: Copy all images into /content/data/panda and /content/data/bear
def safe_copy_images(src_dir, dst_dir):
    copied = 0
    for r, d, f in os.walk(src_dir):
        for fn in f:
            if not is_img(fn):
                continue
            src = os.path.join(r, fn)
            # validate image
            ok = True
            try:
                with Image.open(src) as im:
                    im.verify()
            except Exception:
                ok = False
            if not ok:
                continue
            base, ext = os.path.splitext(os.path.basename(src))
            dst = os.path.join(dst_dir, base + ext)
            i = 1
            while os.path.exists(dst):
                dst = os.path.join(dst_dir, f"{base}_{i}{ext}")
                i += 1
            shutil.copy2(src, dst)
            copied += 1
    return copied

panda_target = "/content/data/panda"
bear_target  = "/content/data/bear"

copied_panda = 0
for p in panda_dirs:
    copied_panda += safe_copy_images(p, panda_target)

copied_bear = 0
for b in bear_dirs:
    copied_bear += safe_copy_images(b, bear_target)

print(f"📦 Copied images → panda: {copied_panda}, bear: {copied_bear}")

# Step 6: Verify the upload
print("\n🎯 Final Folder Structure:")
!find /content/data -type d

print(f"\n📊 Upload Summary:")
print(f"🐼 Panda images: {len([x for x in os.listdir('/content/data/panda') if is_img(x)])}")
print(f"🐻 Bear images: {len([x for x in os.listdir('/content/data/bear') if is_img(x)])}")
print(f"📁 Logs folder: {len(os.listdir('/content/data/logs'))} files")

print("\n✅ Ready for training! Your images are in:")
print("   /content/data/panda/")
print("   /content/data/bear/")

# ==========================
# (Your original flow below)
# ==========================

# GPU memory growth (optional)
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)

tf.config.list_physical_devices('GPU')
len(gpus)

# Image work section - remove incompatible images (already validated by PIL, so this is just a safety pass)
# (kept to mirror your original structure)
data_dir = '/content/data'
image_exts = ['jpeg','jpg', 'bmp', 'png']

# Optional recheck with OpenCV read
removed = 0
for image_class in os.listdir(data_dir):
    class_path = os.path.join(data_dir, image_class)
    if not os.path.isdir(class_path):
        continue
    for image in os.listdir(class_path):
        image_path = os.path.join(class_path, image)
        try:
            if not is_img(image):
                os.remove(image_path); removed += 1
                continue
            img = cv2.imread(image_path)
            if img is None:
                os.remove(image_path); removed += 1
        except Exception:
            try:
                os.remove(image_path); removed += 1
            except:
                pass
print(f"🧹 Removed {removed} incompatible/corrupted files during safety pass.")

# Load Data

# STEP 1: Check current structure
print("📁 Current folder structure:")
!find /content/data -type d

# STEP 2: Remove extra folders (keep only panda and bear)
import shutil as _shutil
folders_to_keep = ['panda', 'bear']

for folder in os.listdir('/content/data'):
    folder_path = os.path.join('/content/data', folder)
    if os.path.isdir(folder_path) and folder not in folders_to_keep and folder != 'logs':
        _shutil.rmtree(folder_path)
        print(f"🗑️ Removed: {folder}")

# STEP 3: Reload dataset
data = tf.keras.utils.image_dataset_from_directory(
    '/content/data',
    class_names=['panda', 'bear'],  # Force 2 classes (0=panda, 1=bear)
    image_size=(256, 256),
    batch_size=32,
    shuffle=True,
    seed=42
)

print(f"🎯 Classes: {data.class_names}")

# STEP 4: Verify labels
batch = data.as_numpy_iterator().next()
images, labels = batch
print(f"📊 Label distribution in batch: {np.unique(labels, return_counts=True)}")

tf.data.Dataset

# Preview a few images
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(f"label={batch[1][idx]} (0=panda,1=bear)")

# - Phase 02 -
# 2.1 Data Scaled (Between 0 & 1)

# scaling to [0,1]
data = data.map(lambda x,y: (tf.cast(x, tf.float32)/255.0, y))

# Get a scaled batch
scaled_iterator = data.as_numpy_iterator()
batch = scaled_iterator.next()
print("Min/Max after scale:", batch[0].min(), batch[0].max())

fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img)
    ax[idx].title.set_text(f"label={batch[1][idx]}")

# 2.2 Split the Data (batch-wise like original)
num_batches = len(data)
train_size = int(num_batches * .7)
val_size   = int(num_batches * .2) + 1
test_size  = int(num_batches * .1) + 1

# ensure they sum to num_batches
if train_size + val_size + test_size > num_batches:
    overflow = train_size + val_size + test_size - num_batches
    test_size = max(1, test_size - overflow)

print("Batches:", num_batches, "→ train/val/test:", train_size, val_size, test_size)

train = data.take(train_size)
val   = data.skip(train_size).take(val_size)
test  = data.skip(train_size + val_size).take(test_size)

# - Phase 03 -
# 3.1 Build Deep Learning Model (kept your original CNN)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout

model = Sequential()
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # binary (0=panda, 1=bear)

model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])

model.summary()

# 3.2 Train (kept your tensorboard callback)
logdir='logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)

hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])

hist.history

# 3.3 Plot Performance

fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()

fig = plt.figure()
if 'accuracy' in hist.history:
    plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
    plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
elif 'acc' in hist.history:
    plt.plot(hist.history['acc'], color='teal', label='accuracy')
    plt.plot(hist.history['val_acc'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.ylim(0, 1.1)
plt.show()

# - Phase 04 -
# Evaluate Performance
# 4.1 Evaluate

from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy

pre = Precision()
re  = Recall()
acc = BinaryAccuracy()

for X, y in test.as_numpy_iterator():
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)

print(pre.result(), re.result(), acc.result())

# 4.2 Test (predict on one image from your dataset)
# We'll pick a random image from /content/data/bear (class 1) to mimic your original single-file test

def pick_random_image(folder):
    cands = [os.path.join(folder, f) for f in os.listdir(folder) if is_img(f)]
    return random.choice(cands) if cands else None

test_img_path = pick_random_image('/content/data/bear') or pick_random_image('/content/data/panda')
print("🖼️ Testing on:", test_img_path)

img = cv2.imread(test_img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb); plt.axis('off'); plt.show()

resize = tf.image.resize(img_rgb, (256,256))
plt.imshow(resize.numpy().astype(int)); plt.axis('off'); plt.show()

yhat = model.predict(np.expand_dims(resize/255, 0))
yhat

if yhat > 0.5:
    print('Predicted class is Bear (1)')
else:
    print('Predicted class is Panda (0)')

# - Phase 05 -
# Save the Model

from tensorflow.keras.models import load_model

os.makedirs('models', exist_ok=True)
model.save(os.path.join('models','panda_bear_classifier.h5'))

new_model = load_model(os.path.join('models','panda_bear_classifier.h5'))
new_model

new_model.predict(np.expand_dims(resize/255, 0))
```
