```python
# =============================================
# PHASE 01: UPLOAD & EXTRACT DATASET
# =============================================
from google.colab import files
import zipfile, os, glob, pandas as pd, numpy as np

print("📤 Please upload your ZIP file that contains the CSV (e.g., 'Email Spam.zip')...")
uploaded = files.upload()  # pick your ZIP from local drive

# Auto-detect uploaded filename (first file)
zip_path = list(uploaded.keys())[0]
print(f"✅ Uploaded file: {zip_path}")

# Extract ZIP to a fixed folder
extract_dir = "/content/email_spam_data"
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("✅ Files extracted to:", extract_dir)
!ls -la "$extract_dir"

# Find a CSV inside the extracted directory
csv_files = glob.glob(os.path.join(extract_dir, "**/*.csv"), recursive=True)
if not csv_files:
    raise FileNotFoundError("No CSV found inside the uploaded ZIP. Please ensure it contains a CSV file.")

csv_path = csv_files[0]
print(f"📄 Using CSV: {csv_path}")

# Try robust CSV loading (handle common encodings)
load_tried = False
encodings_try = [None, "utf-8", "latin-1", "ISO-8859-1"]
for enc in encodings_try:
    try:
        data = pd.read_csv(csv_path, encoding=enc)
        load_tried = True
        break
    except Exception as e:
        last_err = e
if not load_tried:
    raise last_err

print("✅ Loaded CSV with shape:", data.shape)
display(data.head(3))

# =============================================
# PHASE 02: DATA EXPLORATION & CLEANING
# =============================================
print("Columns:", data.columns.tolist())
data.columns = [c.strip().lower() for c in data.columns]  # normalize

# Heuristics to find text/label columns across common spam datasets
# Examples it handles:
#  - 'text'/'label'
#  - 'message'/'spam'
#  - 'email'/'label'
#  - 'v2'/'v1' (SMS Spam Collection)
possible_text_cols = [c for c in data.columns if any(k in c for k in ["text","message","email","content","body","v2"])]
possible_label_cols = [c for c in data.columns if any(k in c for k in ["label","spam","class","target","v1"])]

# Pick first candidate or fallback to first/last column
text_col = possible_text_cols[0] if possible_text_cols else data.columns[0]
label_col = possible_label_cols[0] if possible_label_cols else data.columns[-1]
print(f"📝 Text column: {text_col} | 🏷️ Label column: {label_col}")

# Drop rows with missing text/label
data = data.dropna(subset=[text_col, label_col]).reset_index(drop=True)

# Convert label to binary {0,1}
# Map common variants: 'spam'->1, 'ham'->0, 'not spam'->0, '1'->1, '0'->0
lab_norm = data[label_col].astype(str).str.strip().str.lower()
label_map = {"spam":1, "ham":0, "not spam":0, "not_spam":0, "non-spam":0, "nonspam":0, "1":1, "0":0}
data[label_col] = lab_norm.map(lambda x: label_map[x] if x in label_map else (1 if "spam" in x else 0)).astype(int)

print("Label counts:\n", data[label_col].value_counts(dropna=False))
print("Sample cleaned labels ->", data[label_col].unique())

# =============================================
# PHASE 03: TEXT PREPROCESSING & SPLIT
# =============================================
import re
import tensorflow as tf
from sklearn.model_selection import train_test_split

def clean_text(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", " ", text)   # remove urls
    text = re.sub(r"[^a-zA-Z\s]", " ", text)      # keep letters/space
    text = re.sub(r"\s+", " ", text)              # collapse spaces
    return text.strip()

data[text_col] = data[text_col].astype(str).apply(clean_text)

# Split (stratify = preserve spam/ham ratio)
X_train, X_test, y_train, y_test = train_test_split(
    data[text_col], data[label_col], test_size=0.2, random_state=42, stratify=data[label_col]
)
print("Train size:", len(X_train), "Test size:", len(X_test))

# Tokenize & pad
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = 20000     # increase if you have lots of unique tokens
maxlen = 200           # sequence length; adjust if emails are long

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=maxlen, padding='post', truncating='post')
X_test_seq  = pad_sequences(tokenizer.texts_to_sequences(X_test),  maxlen=maxlen, padding='post', truncating='post')

y_train = np.array(y_train).astype(int)
y_test  = np.array(y_test).astype(int)

print("X_train_seq:", X_train_seq.shape, "X_test_seq:", X_test_seq.shape)

# =============================================
# PHASE 04: BUILD MODEL (LSTM)
# =============================================
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

model = Sequential([
    Embedding(vocab_size, 64, input_length=maxlen),
    LSTM(64, dropout=0.3, recurrent_dropout=0.3),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# =============================================
# PHASE 05: TRAIN
# =============================================
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)
]

history = model.fit(
    X_train_seq, y_train,
    validation_split=0.2,
    epochs=6,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# =============================================
# PHASE 06: EVALUATION
# =============================================
loss, acc = model.evaluate(X_test_seq, y_test, verbose=0)
print(f"\n✅ Test Accuracy: {acc:.4f} | Test Loss: {loss:.4f}")

y_prob = model.predict(X_test_seq, verbose=0).ravel()
y_pred = (y_prob > 0.5).astype("int32")

from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("\n📊 Classification Report:\n", classification_report(y_test, y_pred, digits=4))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Optional: threshold tuning for better recall/precision tradeoff
from sklearn.metrics import precision_recall_curve
prec, rec, thr = precision_recall_curve(y_test, y_prob)
f1 = 2 * (prec * rec) / (prec + rec + 1e-9)
best_idx = int(np.nanargmax(f1))
best_thr = thr[best_idx] if best_idx < len(thr) else 0.5
print(f"🔧 Best threshold by F1: {best_thr:.3f}")

y_pred_tuned = (y_prob > best_thr).astype("int32")
print("\n📊 Tuned threshold report:\n", classification_report(y_test, y_pred_tuned, digits=4))

# =============================================
# PHASE 07: SAVE MODEL & QUICK INFERENCE
# =============================================
os.makedirs("/content/models", exist_ok=True)
save_path = "/content/models/email_spam_classifier.h5"
model.save(save_path)
print("💾 Model saved to:", save_path)

# Quick test
def predict_text(sample: str):
    cleaned = clean_text(sample)
    seq = pad_sequences(tokenizer.texts_to_sequences([cleaned]), maxlen=maxlen, padding='post', truncating='post')
    p = float(model.predict(seq, verbose=0)[0,0])
    cl = "SPAM" if p > best_thr else "NOT SPAM"
    print(f"\n🧪 Text: {sample}\n   Score: {p:.4f}  |  Threshold: {best_thr:.3f}  →  {cl}")

predict_text("Congratulations! You have won a free iPhone. Click here to claim now.")
predict_text("Hi, just checking if we are still on for the meeting tomorrow at 10am.")
```
