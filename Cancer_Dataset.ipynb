```python
# ============================
# Colab: Skin Cancer Classifier
# ============================

# --- Phase 01: Upload & prepare data ---
from google.colab import files
import os, shutil, zipfile, random
from PIL import Image

print("üì§ Upload the ZIP containing the 'Skin_Data' folder...")
uploaded = files.upload()  # select your zipped Skin_Data

# find a zip
zip_names = [k for k in uploaded.keys() if k.lower().endswith(".zip")]
assert len(zip_names) >= 1, "Please upload a .zip file containing Skin_Data/"
zip_path = zip_names[0]
print("‚úÖ ZIP:", zip_path)

# unzip
RAW_DIR = "/content/dataset_raw"
os.makedirs(RAW_DIR, exist_ok=True)
with zipfile.ZipFile(zip_path, "r") as zf:
    zf.extractall(RAW_DIR)
print("‚úÖ Unzipped to:", RAW_DIR)

# locate Skin_Data (supports 'Skin Data' or 'Skin_Data')
def locate_skin_data(root):
    for r, d, f in os.walk(root):
        for name in d:
            norm = name.lower().replace(" ", "_")
            if norm == "skin_data":
                return os.path.join(r, name)
    return None

SKIN_DATA = locate_skin_data(RAW_DIR)
assert SKIN_DATA and os.path.isdir(SKIN_DATA), "Couldn't find 'Skin_Data' inside the ZIP."
print("üìÅ Using:", SKIN_DATA)

# Expected structure:
# Skin_Data/
#   Cancer/
#     Training/
#     Testing/
#   Non_Cancer/
#     Training/
#     Testing/

SRC_TRAIN_C = os.path.join(SKIN_DATA, "Cancer", "Training")
SRC_TEST_C  = os.path.join(SKIN_DATA, "Cancer", "Testing")
SRC_TRAIN_N = os.path.join(SKIN_DATA, "Non_Cancer", "Training")
SRC_TEST_N  = os.path.join(SKIN_DATA, "Non_Cancer", "Testing")

# Target (keep original split)
TRAIN_ROOT = "/content/data_train"
TEST_ROOT  = "/content/data_test"
os.makedirs(os.path.join(TRAIN_ROOT, "cancerous"), exist_ok=True)
os.makedirs(os.path.join(TRAIN_ROOT, "non_cancerous"), exist_ok=True)
os.makedirs(os.path.join(TEST_ROOT,  "cancerous"), exist_ok=True)
os.makedirs(os.path.join(TEST_ROOT,  "non_cancerous"), exist_ok=True)

# copy helper with simple image validation via PIL (avoid imghdr)
VALID_EXTS = (".jpg",".jpeg",".png",".bmp",".tif",".tiff",".webp")

def safe_copy_images(src_dir, dst_dir):
    count, removed = 0, 0
    if not os.path.isdir(src_dir):
        return 0, 0
    for r, d, f in os.walk(src_dir):
        for fn in f:
            if not fn.lower().endswith(VALID_EXTS):
                continue
            src = os.path.join(r, fn)
            # validate
            ok = True
            try:
                with Image.open(src) as im:
                    im.verify()  # detect truncated/corrupt
            except Exception:
                ok = False
            if not ok:
                removed += 1
                continue
            # unique name if collision
            base, ext = os.path.splitext(os.path.basename(src))
            dst = os.path.join(dst_dir, base + ext)
            i = 1
            while os.path.exists(dst):
                dst = os.path.join(dst_dir, f"{base}_{i}{ext}")
                i += 1
            shutil.copy2(src, dst)
            count += 1
    return count, removed

c_tr, c_bad = safe_copy_images(SRC_TRAIN_C, os.path.join(TRAIN_ROOT, "cancerous"))
c_te, c_bad2 = safe_copy_images(SRC_TEST_C,  os.path.join(TEST_ROOT,  "cancerous"))
n_tr, n_bad = safe_copy_images(SRC_TRAIN_N, os.path.join(TRAIN_ROOT, "non_cancerous"))
n_te, n_bad2 = safe_copy_images(SRC_TEST_N,  os.path.join(TEST_ROOT,  "non_cancerous"))

print(f"üì¶ Train ‚Üí cancerous: {c_tr}, non_cancerous: {n_tr} (skipped bad: {c_bad+n_bad})")
print(f"üì¶ Test  ‚Üí cancerous: {c_te}, non_cancerous: {n_te} (skipped bad: {c_bad2+n_bad2})")

def count_files(p):
    try:
        return len([f for f in os.listdir(p) if os.path.isfile(os.path.join(p,f))])
    except FileNotFoundError:
        return 0

print("\nüìä Final counts")
print("Train cancerous:", count_files(os.path.join(TRAIN_ROOT,"cancerous")))
print("Train non_cancerous:", count_files(os.path.join(TRAIN_ROOT,"non_cancerous")))
print("Test  cancerous:", count_files(os.path.join(TEST_ROOT,"cancerous")))
print("Test  non_cancerous:", count_files(os.path.join(TEST_ROOT,"non_cancerous")))

# --- Phase 02: Build datasets (train/val from training; test from testing) ---
import tensorflow as tf
from tensorflow.keras import layers

# GPU memory growth (optional)
gpus = tf.config.experimental.list_physical_devices("GPU")
for g in gpus:
    tf.config.experimental.set_memory_growth(g, True)
print("GPUs:", tf.config.list_physical_devices("GPU"))

IMG_SIZE   = (224, 224)        # EfficientNetB0 default
BATCH_SIZE = 32
SEED       = 42

# Create train dataset first (we'll carve a val split from it)
train_all = tf.keras.utils.image_dataset_from_directory(
    TRAIN_ROOT,
    class_names=['cancerous','non_cancerous'],  # fixes label order: 0=cancerous, 1=non_cancerous
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

# A small validation split taken from the end of shuffled train batches
num_batches = tf.data.experimental.cardinality(train_all).numpy()
val_batches = max(1, int(num_batches * 0.2))
train_batches = max(1, num_batches - val_batches)
train = train_all.take(train_batches)
val   = train_all.skip(train_batches).take(val_batches)

# Held-out test from the dataset's original Testing folders
test = tf.keras.utils.image_dataset_from_directory(
    TEST_ROOT,
    class_names=['cancerous','non_cancerous'],
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

# Data augmentation only on training
augment = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="augment")

# Normalize to [0,1] (EfficientNet will still work fine with this rescaling)
def prep(ds, training=False):
    ds = ds.map(lambda x,y: (tf.cast(x, tf.float32)/255.0, y),
                num_parallel_calls=tf.data.AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (augment(x, training=True), y),
                    num_parallel_calls=tf.data.AUTOTUNE)
    return ds.cache().prefetch(tf.data.AUTOTUNE)

train = prep(train, training=True)
val   = prep(val, training=False)
test  = prep(test, training=False)

# --- Phase 03: Model (EfficientNetB0 transfer learning) ---
from tensorflow.keras import Model, Input
from tensorflow.keras.applications import EfficientNetB0

base = EfficientNetB0(include_top=False, weights="imagenet",
                      input_shape=(*IMG_SIZE, 3), pooling="avg")
base.trainable = False  # start frozen

inp = Input(shape=(*IMG_SIZE, 3))
x = layers.Rescaling(1.0)(inp)   # already [0,1] via dataset map; keep as identity to be explicit
x = base(x, training=False)
x = layers.Dropout(0.2)(x)
out = layers.Dense(1, activation='sigmoid')(x)
model = Model(inp, out)

model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])

model.summary()

# --- Phase 03.1: Class weights (handle imbalance) ---
import math

n_pos = count_files(os.path.join(TRAIN_ROOT,"cancerous"))
n_neg = count_files(os.path.join(TRAIN_ROOT,"non_cancerous"))
total = max(1, n_pos + n_neg)
# Labels: 0=cancerous, 1=non_cancerous (because of class_names order)
class_weight = {
    0: total / (2.0 * max(1, n_pos)),
    1: total / (2.0 * max(1, n_neg)),
}
print("class_weight:", class_weight)

# --- Phase 03.2: Train with callbacks ---
es  = tf.keras.callbacks.EarlyStopping(monitor='val_auc', mode='max', patience=4, restore_best_weights=True)
rlr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.25, patience=2, min_lr=1e-6)
tb  = tf.keras.callbacks.TensorBoard(log_dir="logs")

history = model.fit(
    train,
    validation_data=val,
    epochs=20,
    class_weight=class_weight,
    callbacks=[es, rlr, tb],
    verbose=1
)

# Optionally unfreeze top layers for fine-tuning (if val AUC plateaus but not great)
# base.trainable = True
# model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
#               loss=tf.keras.losses.BinaryCrossentropy(),
#               metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
# history_ft = model.fit(train, validation_data=val, epochs=5,
#                        class_weight=class_weight, callbacks=[es, rlr, tb])

# --- Phase 03.3: Curves ---
import matplotlib.pyplot as plt

plt.figure(); plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss'); plt.legend(); plt.title('Loss'); plt.show()

acc_key = 'accuracy' if 'accuracy' in history.history else 'acc'
v_acc_key = 'val_accuracy' if 'val_accuracy' in history.history else 'val_acc'
plt.figure(); plt.plot(history.history[acc_key], label='accuracy')
plt.plot(history.history[v_acc_key], label='val_accuracy'); plt.ylim(0,1.05); plt.legend(); plt.title('Accuracy'); plt.show()

plt.figure(); plt.plot(history.history['auc'], label='auc')
plt.plot(history.history['val_auc'], label='val_auc'); plt.ylim(0,1.05); plt.legend(); plt.title('AUC'); plt.show()

# --- Phase 04: Tune threshold on validation set (maximize Youden's J) ---
import numpy as np
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report

y_true_val, y_score_val = [], []
for Xb, yb in val:
    p = model.predict(Xb, verbose=0).ravel()
    y_score_val.extend(p.tolist())
    y_true_val.extend(yb.numpy().tolist())
y_true_val = np.array(y_true_val)
y_score_val = np.array(y_score_val)

fpr, tpr, thr = roc_curve(y_true_val, y_score_val)
youden = tpr - fpr
best_idx = int(np.argmax(youden))
best_thr = float(thr[best_idx])
print(f"\nüîß Best threshold (Youden): {best_thr:.4f} | Val AUC: {auc(fpr,tpr):.4f}")

# --- Phase 04.1: Evaluate on TEST at 0.5 vs tuned threshold ---
def collect_scores(ds):
    y_true, y_score = [], []
    for Xb, yb in ds:
        p = model.predict(Xb, verbose=0).ravel()
        y_score.extend(p.tolist())
        y_true.extend(yb.numpy().tolist())
    return np.array(y_true), np.array(y_score)

y_true_t, y_score_t = collect_scores(test)

def evaluate_at(thr, name):
    y_pred = (y_score_t >= thr).astype(int)
    print(f"\n=== Test @ {name} (thr={thr:.3f}) ===")
    print(confusion_matrix(y_true_t, y_pred))
    print(classification_report(y_true_t, y_pred,
          target_names=['cancerous (0)','non_cancerous (1)'], digits=4))

evaluate_at(0.5, "default")
evaluate_at(best_thr, "tuned")

# --- Phase 04.2: Demo prediction on a random image ---
def pick_random_image(dir_path):
    cands = []
    for r, d, f in os.walk(dir_path):
        for fn in f:
            if fn.lower().endswith(VALID_EXTS):
                cands.append(os.path.join(r, fn))
    return random.choice(cands) if cands else None

demo_path = pick_random_image(os.path.join(TEST_ROOT, "cancerous"))
if demo_path is None:
    demo_path = pick_random_image(os.path.join(TEST_ROOT, "non_cancerous"))

print("\nüñºÔ∏è Demo image:", demo_path)
img = Image.open(demo_path).convert("RGB").resize(IMG_SIZE)
arr = np.array(img).astype("float32") / 255.0
score = float(model.predict(arr[None, ...], verbose=0)[0,0])
print(f"Raw score (sigmoid): {score:.6f}")
print("Pred @0.5:       ", "cancerous" if score >= 0.5 else "non_cancerous")
print("Pred @best_thr:  ", "cancerous" if score >= best_thr else "non_cancerous")

plt.imshow(img); plt.axis('off'); plt.show()

# --- Phase 05: Save model ---
os.makedirs("models", exist_ok=True)
keras_path = "models/skin_cancer_classifier.keras"   # native Keras format
model.save(keras_path)
print("üíæ Saved:", keras_path)
```
